````python
response = llm.create_chat_completion(
    messages=[
        {
            "role": "system",
            "content": "あなたは誠実で優秀な日本人のアシスタントです。特に指示が無い場合は、常に日本語で回答してください。",
        },
        {
            "role": "user",
            "content": f"以下の文章を翻訳してください。\n {text}",
        },
    ],
    max_tokens=1024,
)

print(response["choices"][0]["message"]["content"])
````
